# Linear Regression with Tensorflow

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

### Creating the dataset

```python
m = 200
t0 = np.random.randn()
t1 = np.random.randn()
x = np.random.normal(0.0,0.5,m)
y = t0 + x * t1 + np.random.normal(0.0,1.0,m) / 10
plt.scatter(x,y,alpha=0.4)
plt.show()
```

### Initialize the variables

```python
theta = tf.Variable(tf.random_uniform([2], -1.0, 1.0))
```

### Build the model

```python
yhat = theta[0] + theta[1] * x

learning_rate = 0.5
iterations = 25

J = tf.reduce_mean(tf.square(yhat-y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train = optimizer.minimize(J) 
```

### Training

```python
model = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(model)
    
    for i in range(iterations):
        sess.run(train)
        
        if i%5 == 0:
            plt.scatter(x,y,alpha=0.4)
            plt.plot(x,sess.run(theta[1])*x+sess.run(theta[0]))
            plt.legend()
            plt.show()
```
